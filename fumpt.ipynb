# Explorando dados com formatos em Go

Go é uma linguagem mais próxima da máquina, como C. 
Tem tipos como `byte`, `uint16`, `float64` que são tipos
nativos nas CPUs.
O tipo `string` em Go é explicitamente uma sequência
de bytes, mas com facilidades para representar texto em Unicode.
Go também tem tipos compostos como
_slice_, _struct_, e _map_.

Para explorar toda esta variedade de tipos, vale a pena conhecer
a linguagem de formatação do pacote `fmt` (pronúncia tradicional: _fumpt_),
que permite exibir dados de todos os tipos de várias formas diferentes.

Vamos começar com o tipo `rune` (runa), a versão Go do `char` em C ou Java.

O código abaixo cria uma runa e a exibe de várias formas:

```go
var cat = '\U0001F638'
%%
fmt.Printf("%v | %x | %c | %q | %U", cat, cat, cat, cat, cat)
```

A documentação do [pacote `fmt`](https://pkg.go.dev/fmt) chama de _verbs_ (verbos)
os códigos de formatação como `%v`, `%q` e `%U`.

No exemplo temos os seguintes verbos:

* `%v`: o formato padrão (_default_) daquele tipo de valor; no caso de `rune`, é um número decimal;
* `%x`: o número apresentado em hexadecimal;
* `%c`: o caractere identificado pela runa;
* `%q`: o caractere entre aspas simples;
* `%U`: o código hexadecimal no formato padrão do Unicode

Alguns verbos aceitam modificadores, como `#` no exemplo abaixo:


```go
%%
fmt.Printf("%x | %#x | %U | %#U\n", cat, cat, cat, cat)
```


Dois verbos que servem para qualquer tipo são `%v` e `%T`:

```go
var drink = "café"
%%
fmt.Printf("%v  %#v  %T", drink, drink, drink)
```

Algums verbos podem ser modificados com opções, como `#` em `%#v` ou `+` em `%+d`;
são chamadas _flags_ na
[documentação](https://pkg.go.dev/fmt).

## Verbos para `rune` e `string`

```go
var drink = "café 🥤!"
%%
fmt.Printf("%q %+q", drink, drink)
```

A opção `%+q` garante saída só com caracteres ASCII,
substituindo qualquer outro por um escape Unicode de
4 dígitos (`\u....`) ou 8 dígitos (`\U.........`),
conforme a necessidade.

Para runas, o verbo `%c` mostra o caractere, sem aspas.

```go
var ohm = '\u2126'
%%
fmt.Printf("%q %+q %c", ohm, ohm, ohm)
```

Os verbos `%q` e `%c` não se limitam ao tipo `rune`:
exibem qualquer tipo inteiro como um caractere Unicode.

```go
var piu int16 = 0x3c0
%%
fmt.Printf("%q %+q %c", piu, piu, piu)
```

```go
var atilde byte = 227
%%
fmt.Printf("%q %+q %c", atilde, atilde, atilde)
```

```go
var nãoExiste rune = 0xdead
%%
fmt.Printf("%q %+q %c", nãoExiste, nãoExiste, nãoExiste)
```

```go
var i int = 42
var f32 float32 = math.Pi
var f64 float64 = math.Pi
var runa = 'A'
var str = "A"
var primos = []int{2, 3, 5, 7, 11, 13}
var sabores = []string{"banana", "açaí", "uba"}
```

O verbo `%v` especifica o _default format_ (formato padrão). 

As funções `fmt.Print` e `fmt.Println` usam o formado `%v` automaticamente.

```go
%%
fmt.Println(i)
fmt.Println(f32)
fmt.Println(f64)
fmt.Println(runa)
fmt.Println(str)
fmt.Println(primos)
fmt.Println(sabores)
```

A opção `#v` exibe o valor na sintaxe literal de Go:

```go
%%
fmt.Printf("%#v\n", str)
fmt.Printf("%#v\n", primos)
fmt.Printf("%#v\n", sabores)
```

```go
type Livro struct {
		Título string
		Ano  int
	}
var livro = Livro{"O Alienista", 1882}
%%
fmt.Printf("%v\n", livro)
fmt.Printf("%#v\n", livro)
```

```go
%%
conto := Livro{Ano:1882, Título:"O Alienista"}
fmt.Printf("%#v\n", conto)
fmt.Print(conto == livro)
```

```go
%%
resposta := 42

fmt.Println(resposta)
fmt.Printf("%v\n", resposta)
fmt.Printf("%d\n", resposta)
```

The special verb `%T` shows the type of an item rather than its value.
