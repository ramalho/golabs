```go
var (
    n = -1
    i   int
    u8  uint8
    i32 int32
    u64 uint64
)

%%
fmt.Printf("%T\t%v\t%v\n", n, n, n-1);
fmt.Printf("%T\t%v\t%v\n", i, i, i-1);
fmt.Printf("%T\t%v\t%v\n", u8, u8, u8-1);
fmt.Printf("%T\t%v\t%v\n", i32, i32, i32-1);
fmt.Printf("%T\t%v\t%v\n", u64, u64, u64-1);
```

```go
%%
var m uint64 = math.MaxUint64
fmt.Printf("%T\t%v\n", m, m);
```
